## Numpy basics

```python
import numpy as np

# 1D array
arr_1d = np.array([1, 2, 3, 4])

# 2D array
arr_2d = np.array([[1, 2], [3, 4]])

# Zeros, ones, arange, and linspace
zeros_array = np.zeros((3, 3))
ones_array = np.ones((2, 4))
range_array = np.arange(0, 10, 2)
linspace_array = np.linspace(0, 1, 5)
```

## Slicing

```python
arr = np.array([1, 2, 3, 4])

# Indexing and Slicing
print(arr[0])  # First element
print(arr[1:3])  # Slice from index 1 to 2

# Mathematical Operations
print(arr + 2)
print(arr * 3)
print(np.sqrt(arr))
```

## Operations

```python
reshaped_arr = np.arange(1, 10).reshape(3, 3)
print(reshaped_arr)
print("----------------------------------------")
transposed_arr = reshaped_arr.T
print(transposed_arr)
```

```python
arr1 = np.array([1, 2, 3])
print(arr1)
print("-----------------------------------------")
arr2 = np.array([[10], [20], [30]])
print(arr2)
print("-----------------------------------------")
broadcasted_result = arr1 + arr2
print(broadcasted_result)
```

## Matrix Operations

```python
# 5x5 matrix with border elements as 0
ones_matrix = matrix = np.ones((5, 5))
print(ones_matrix)
print("-----------------------------------------")
matrix[1:-1, 1:-1] = 0
print(matrix)
# Random 10x10 matrix, mean, median, and std
rand_matrix = np.random.rand(10, 10)
print(rand_matrix)
print("-----------------------------------------")
mean_val = np.mean(rand_matrix)
print(mean_val)
print("-----------------------------------------")
median_val = np.median(rand_matrix)
print(median_val)
print("-----------------------------------------")
std_val = np.std(rand_matrix)
print(std_val)
print("-----------------------------------------")

# 3x3 identity matrix
identity_matrix = np.eye(3)
print(identity_matrix)
print("-----------------------------------------")

# Matrix multiplication
mat1 = np.array([[1, 2], [3, 4]])
print(mat1)
print("-----------------------------------------")
mat2 = np.array([[5, 6], [7, 8]])
print(mat2)
print("-----------------------------------------")
dot_product = np.dot(mat1, mat2)
print(dot_product)
print("-----------------------------------------")

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(mat1)
print(eigenvalues)
print("-----------------------------------------")
print(eigenvectors)
print("-----------------------------------------")

# Inverse of a matrix
inverse_mat = np.linalg.inv(mat1)
print(inverse_mat)
print("-----------------------------------------")
```

## Pandas

```python
import pandas as pd

# Create Series and DataFrame
series = pd.Series([1, 2, 3, 4])
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df
# Load CSV (uncomment if you have a CSV file)
# df_csv = pd.read_csv('file.csv')
```

```python
# Selecting columns
age_column = df['Age']

# Filtering rows
filtered_df = df[df['Age'] > 25]
print(filtered_df)
# Indexing with loc and iloc
loc_df = df.loc[0]  # Select first row
print(loc_df)
print("-----------------------------------------")
iloc_df = df.iloc[1:3]  # Select rows by index
print(iloc_df)
```

## Handling missing data

```python
# Handling missing data
df_with_nan = df.copy()
df_with_nan.loc[1, 'Age'] = None
print(df_with_nan)
print("-----------------------------------------")
# Drop missing data
cleaned_df = df_with_nan.dropna()
print(cleaned_df)
print("-----------------------------------------")
# Fill missing data
filled_df = df_with_nan.fillna(df_with_nan['Age'].mean())
print(filled_df)
```

## Aggregation Functions

```python
# Aggregation functions
sum_column = df['Age'].sum()
print(sum_column)
print("-----------------------------------------")
mean_column = df['Age'].mean()
print(mean_column)
print("-----------------------------------------")
max_column = df['Age'].max()
print(max_column)
print("-----------------------------------------")
min_column = df['Age'].min()
print(min_column)
```

## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset and scatter plot
tips = sns.load_dataset("tips")
print(tips.head())
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()

# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
print(tips.dtypes)
numeric_tips = tips.select_dtypes(include=[np.number])
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Fill NaN values with 0 or any other default value
tips['sex'] = tips['sex'].fillna('Male')
tips['smoker'] = tips['smoker'].fillna('No')

# Then apply the mapping
tips['sex'] = tips['sex'].map({'Male': 1, 'Female': 0})
tips['smoker'] = tips['smoker'].map({'Yes': 1, 'No': 0})
```

## Basics of Plotting

```python
import matplotlib.pyplot as plt
import numpy as np

# Basic Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.show()
```

## Subplots

```python
# Create multiple plots in the same figure
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Sine')
plt.title("Sine Wave")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='r', label='Cosine')
plt.title("Cosine Wave")

plt.tight_layout()
plt.show()
```

## Customizing plots

```python
# Customize the plot with colors, markers, and labels
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Sine')

plt.title("Customized Sine Wave")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.grid(True)
plt.show()
```

## Bar Charts

```python
# Customize the plot with colors, markers, and labels
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Sine')

plt.title("Customized Sine Wave")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
# Scatter plot with random data
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', marker='x')
plt.title("Random Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

## Histogram

```python
# Histogram of normally distributed data
data = np.random.randn(1000)

plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
